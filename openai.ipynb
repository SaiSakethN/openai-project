```python
from langchain_core.prompts import ChatPromptTemplate

prompt=ChatPromptTemplate.from_messages(
    [
        ("system","you are an expert in Artificial intelligence. provide me answers based on the questions"),
        ("user","{input}")
    ]
)
prompt
```

```python
chain=prompt|llm

response=chain.invoke({"input":"can you tell me about llm"})
print(response.content)
```

```python
from langchain_core.output_parsers import StrOutputParser
output_parser=StrOutputParser()
chain=prompt|llm|output_parser

response=chain.invoke({"input":"can you tell me about rag"})
print(response)
```
